```python
import json
import pandas as pd
```

# Yelp Reviews

```python
businesses = [] #this is actually review = []
for line in open('review.json', 'r'):
    businesses.append(json.loads(line))
```

```python
# review df
df = pd.DataFrame(businesses)
```

```python
df
```

```python
# save reviews data into csv
df.to_csv('Yelp_Dataset_Reviews.csv')
```
